```python
import pandas as pd
df = pd.read_csv("dataset_with_estimations_exact_annealer_results.csv")
df
```

```python
# For MINLP:
df_minlp = df[~df["is_valid_MINLP"].isna()]
print(f"MINLP dataset len (lower due to errors in optimizer): {len(df_minlp)}")
MINLP_share_is_valid = df_minlp["is_valid_MINLP"].value_counts()[True]*100/len(df_minlp)
MINLP_min_energy_from_valid = df_minlp[df_minlp["is_valid_MINLP"]]["min_energy_MINLP"].min()
print(f"MINLP_share_is_valid: {MINLP_share_is_valid}% MINLP_min_energy_from_valid: {MINLP_min_energy_from_valid}")
```

```python
# For Uniform:
df_uniform_annealer = df[~df["is_valid_uniform_annealer"].isna()]
print(f"Uniform annealer dataset len: {len(df_uniform_annealer)}")
Uniform_annealer_share_is_valid = df_uniform_annealer["is_valid_uniform_annealer"].value_counts()[True]*100/len(df_uniform_annealer)
Uniform_annealer_min_energy_from_valid = df_uniform_annealer[df_uniform_annealer["is_valid_uniform_annealer"]]["min_energy_uniform_annealer"].min()
print(f"Uniform_annealer_share_is_valid: {Uniform_annealer_share_is_valid}, Uniform_annealer_min_energy_from_valid: {Uniform_annealer_min_energy_from_valid}")
```

```python
# For ML estimated exact annealer:
df_exact_annealer = df[~df["is_valid_exact"].isna()]
print(f"Exact annealer dataset len: {len(df_exact_annealer)}")
exact_annealer_share_is_valid = df_exact_annealer["is_valid_exact"].value_counts()[True]*100/len(df_exact_annealer)
exact_annealer_min_energy_from_valid = df_exact_annealer[df_exact_annealer["is_valid_exact"]]["min_energy_exact"].min()
print(f"exact_annealer_share_is_valid: {exact_annealer_share_is_valid}, exact_annealer_min_energy_from_valid: {exact_annealer_min_energy_from_valid}")
```

```python
# For uniform quatum annealer:
df_quantum_annealer = df[~df["is_valid_hybrid"].isna()]
print(f"Quantum annealer dataset len: {len(df_quantum_annealer)}")
hybrid_annealer_share_is_valid = df_quantum_annealer["is_valid_hybrid"].value_counts()[True]*100/len(df_quantum_annealer)
hybrid_annealer_min_energy_from_valid = df_quantum_annealer[df_quantum_annealer["is_valid_hybrid"]]["min_energy_exact"].min()
print(f"hybrid_annealer_share_is_valid: {hybrid_annealer_share_is_valid}, hybrid_annealer_min_energy_from_valid: {hybrid_annealer_min_energy_from_valid}")
```
